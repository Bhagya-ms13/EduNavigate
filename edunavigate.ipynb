```python
skills_split = courses['Skills'].str.split(',', expand=True)
```

```python
skills_split.dropna(axis=1, how='any', inplace=True)
```

```python
skills_split.shape
```

```python
for i in range(skills_split.shape[1]):
    skills_split.rename(columns={i: f"skill{i+1}"}, inplace=True)
```

```python
courses = pd.concat([courses, skills_split], axis=1)
```

```python
courses.head()
```

```python
courses.drop_duplicates(inplace=True)
```

```python
courses['combined_skills'] = courses.apply(lambda row: [row['skill1'], row['skill2'], row['skill3']], axis=1)
```

```python
# Allow the person to input their top 3 skills
top_skills = []
for i in range(3):
    skill = input(f"Enter skill {i+1}: ")
    top_skills.append(skill)

print("Top skills entered by the person:", top_skills)
```

```python
# User's input of top skills (example)
top_skills = ['Python', 'Data Structures', 'Algorithms']
user_skills_str = ', '.join(top_skills)

# Add the user's skills to the DataFrame
courses.loc[-1] = ['User', 'User', user_skills_str, top_skills[0], top_skills[1], top_skills[2], user_skills_str]

# Use CountVectorizer to convert text into a matrix of token counts
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(courses['combined_skills'])

# Compute cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)

# Get the index of the user row
user_index = courses[courses['Category'] == 'User'].index[0]

# Get cosine similarity values for the user row
similarity_scores = list(enumerate(cosine_sim[user_index]))

# Sort the similarity scores
similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

# Get the top 5 similar courses (excluding the user)
similar_courses = similarity_scores[1:6]

# Get the indices of the top 5 similar courses
similar_course_indices = [x[0] for x in similar_courses]

# Get the corresponding category and subcategory
recommended_categories = courses.loc[similar_course_indices, 'Category']
recommended_subcategories = courses.loc[similar_course_indices, 'Sub-Category']

# Print the recommendations
print("Recommended Categories:")
print(recommended_categories)
print("\nRecommended Subcategories:")
print(recommended_subcategories)
```
